# KNN classification on animals classes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import random
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
with open("/kaggle/input/animal-categories-90-masters-of-survival/name of the animals.txt",'r') as f:
    animals_category = f.read().split("\n")
```

```python
X = []
y = []

for animal_category in animals_category:
    category_images_name = os.listdir("/kaggle/input/animal-categories-90-masters-of-survival/animals/animals/animals/"+animal_category)
    for im_name in category_images_name:
        im_arr = cv2.imread("/kaggle/input/animal-categories-90-masters-of-survival/animals/animals/animals/"+animal_category+"/"+im_name,cv2.IMREAD_GRAYSCALE)
        im_arr = cv2.resize(im_arr,(265,265))
        im_arr = im_arr.flatten()
        
        X.append(im_arr)
        y.append(animal_category)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=123)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

### So KNN is not a good algorithm for classification in this case
